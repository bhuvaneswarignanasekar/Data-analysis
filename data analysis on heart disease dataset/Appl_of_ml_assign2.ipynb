```python
import pandas as pd
df=pd.read_excel('/content/Heart Disease.xlsx')
df.describe(include='all')
```

# Question 1: Statistics

```python
n_at_median, n_at_mode, mode, nzero, nmissing=[],[],[],[],[]
stat=pd.DataFrame()
stat['cardinality']=df.nunique()
stat['mean']=df.mean()
stat['median']=df.median()
stat['stddev']=df.std()
stat['min']=df.min()
stat['max']=df.max()
median=list(stat['median'])
columns=list(df.columns)
j=0
for i in columns:
  m=median[j]
  mo=df[i].mode().iat[0]

  mode.append(mo)
  n_at_median.append((df[i]==m).sum())
  n_at_mode.append((df[i]==mo).sum())
  nzero.append((df[i]==0).sum())
  nmissing.append(df[i].isnull().sum())

  j+=1

stat.insert(3,'n_at_median', n_at_median)
stat.insert(4,'mode', mode) 
stat.insert(5,'n_at_mode', n_at_mode) 
stat['nzero']=nzero
stat['nmissing']=nmissing
s=['cardinality','mean','median','n_at_median','mode','n_at_mode','stddev','min','max','nzero','nmissing']
stat_tran = stat.T
stat_t=stat_tran
stat_t.insert(0,'stat', s) 
stat_t.to_excel('/content/statistics.xlsx', index = False)
stat_t
```

# Question 3 b): missing values

```python
for i in columns:
  print(i,'-',df[i].isnull().sum(),' missing')
```

# Question 3 d): outliers

```python
# outliers
col=['member', 'age', 'cp', 'trestbps', 'chol', 'fbs', 'restecg', 'thalach', 'exang', 'oldpeak', 'slope', 'ca', 'thal', 'target']
for c in col:
  q1=df[c].quantile(0.25)
  q3=df[c].quantile(0.75)
  iqr=q3-q1
  l_lim=q1-(1.5*iqr)
  u_lim=q3+(1.5*iqr)
  mem=list(df[c])
  o_l=[i for i in mem if i<l_lim]
  u_l=[i for i in mem if i>u_lim]
  print("outliers in ",c," : ",len(o_l)+len(u_l))
```

# Question 4: Correlation and Covariance Matrix

```python
cov=pd.DataFrame.cov(df)
cor=df.corr()
cov.insert(0,'column', col) 
cor.insert(0,'column', col) 
cov.to_excel('/content/cov.xlsx', index = False)
cor.to_excel('/content/cor.xlsx', index = False)
```
